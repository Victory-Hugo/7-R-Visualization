```python
import pandas as pd
import matplotlib.pyplot as plt
from aquarel import load_theme
```

```python
df_20K= pd.read_csv('C:/Users/victo/Desktop/Affey_chip.csv')

# 绘图的主题
theme = load_theme('boxy_light') #['arctic_dark', 'arctic_light', 'boxy_dark', 'boxy_light', 'gruvbox_dark', 'gruvbox_light', 'minimal_dark', 'minimal_light', 'scientific', 'umbra_dark', 'umbra_light']
theme.apply()
df_20K
```

```python
# 将20K数据集的Population_City列进行分割，取后者的值
df_20K['City'] = df_20K['Population_City'].str.split('_').str[-1]
df_20K
```

```python
# 排除汉族之后再可视化
df_20K = df_20K.loc[df_20K['Population']!="Han"]
df_20K
```

```python
info_list = ['Province','Population','City','Haplogroup_PCA','Language','Haplogroup_inhouse']
```

```python
for info in info_list:
	# 先按要求分组和计数
	df_20K_INFO = df_20K.groupby([info]).count().loc[:, ['ID']].reset_index()

	# 重命名列名
	df_20K_INFO.columns = ['Haplogroup', 'N']

	# 将 DataFrame 导出为 .txt 文件
	df_20K_INFO.to_csv('C:/Users/victo/Desktop/新建 Text Document.txt', sep='\t', index=False)

	# df_20K_INFO 是重命名后的最终结果


	# 设置plt.rcParams以确保输出为矢量图格式并且字体为 Arial
	plt.rcParams['pdf.fonttype'] = 42
	plt.rcParams['ps.fonttype'] = 42
	plt.rcParams['font.family'] = 'arial'


	# 读取CSV文件的函数
	def load_data(file_path):
		return pd.read_csv(file_path, sep="\t")

	# 准备前30个类别的数据，并将其他类别归纳为“Other”
	classification_numbers = 30
	classification__color_numbers = classification_numbers

	def prepare_top_10_data(df, category_col, value_col):
		# 按数量进行降序排序，并保留前30名，其他类别汇总为“Other”
		df_sorted = df.sort_values(by=value_col, ascending=False)
		top_10 = df_sorted.iloc[:classification_numbers]
		if len(df_sorted) > classification_numbers:
			other = pd.DataFrame({category_col: ['Other'], value_col: [df_sorted.iloc[classification_numbers:][value_col].sum()]})
			top_10 = pd.concat([top_10, other], ignore_index=True)
		return top_10

	# 创建包含柱状图和饼图的合并图片，不含柱状图内部的线条，并保存为矢量图格式
	def create_combined_charts_no_lines(df, category_col, value_col, title_prefix, output_file):
		# 使用高级SCI渐变色配色
		cmap = plt.get_cmap('viridis') # 备选项：plasma，inferno，magma，cividis，viridis，YlGnBu_
		colors = [cmap(i / classification__color_numbers) for i in range(len(df))]

		# 创建两个子图，分别放置柱状图和饼图
		fig, axs = plt.subplots(1, 2, figsize=(14, 7))

		# 在第一个子图中创建柱状图
		bars = axs[0].bar(df[category_col], df[value_col], color=colors)
		#axs[0].set_title(f'{title_prefix} - bar', fontsize=14)
		axs[0].tick_params(axis='x', rotation=90)
		axs[0].set_ylabel('Number', fontsize=12)

		# 在每个柱子上显示数值（Number）
		for bar in bars:
			height = bar.get_height()
			axs[0].text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom', fontsize=7)

		# 移除多余的背景元素，如网格线和轴线
		axs[0].grid(False)
		# axs[0].spines['top'].set_visible(False)
		# axs[0].spines['right'].set_visible(False)
		# axs[0].spines['left'].set_visible(False)
		# axs[0].spines['bottom'].set_visible(False)

		# 在第二个子图中创建饼图，并将比例显示在类别名后的括号内
		total_value = df[value_col].sum()
		labels_with_pct = [f'{category} ({value/total_value:.1%})' for category, value in zip(df[category_col], df[value_col])]
		
		axs[1].pie(df[value_col], labels=labels_with_pct, colors=colors, startangle=140, textprops={'fontsize': 7})
		#axs[1].set_title(f'{title_prefix} - pie', fontsize=14)
		theme.apply_transforms()
		# 调整布局，使图表不重叠
		plt.savefig(f'C:/Users/victo/Desktop/{info}分布图.pdf')
		plt.tight_layout()
		plt.show()

	# 加载所有文件数据，包括新的 4.csv 文件
	file_1 = load_data('C:/Users/victo/Desktop/新建 Text Document.txt')
	# file_2 = load_data('C:/Users/victo/Desktop/2.csv')
	# file_3 = load_data('C:/Users/victo/Desktop/3.csv')
	# file_4 = load_data('C:/Users/victo/Desktop/4.csv')  # 新增的文件
	# file_5 = load_data('C:/Users/victo/Desktop/5.csv')

	# 准备前30个类别的数据
	file_1_prepared = prepare_top_10_data(file_1, 'Haplogroup', 'N')
	# file_2_prepared = prepare_top_10_data(file_2, 'Haplogroup', 'N')
	# file_3_prepared = prepare_top_10_data(file_3, 'Haplogroup', 'N')
	# file_4_prepared = prepare_top_10_data(file_4, 'Language', 'N')  # 新增的语言数据
	# file_5_prepared = prepare_top_10_data(file_5, 'Ethnicity', 'N')

	# 生成并保存可视化图表为SVG文件
	create_combined_charts_no_lines(file_1_prepared, 'Haplogroup', 'N', 'Haplogroup', 'haplogroup_chart.svg')
	# create_combined_charts_no_lines(file_2_prepared, 'Haplogroup', 'N', 'Haplogroup', 'haplogroup_chart_2.svg')
	# create_combined_charts_no_lines(file_3_prepared, 'Haplogroup', 'N', 'Haplogroup', 'haplogroup_chart_3.svg')
	# create_combined_charts_no_lines(file_4_prepared, 'Language', 'N', 'Language', 'language_chart.svg')
	# create_combined_charts_no_lines(file_5_prepared, 'Ethnicity', 'N', 'Ethnicity', 'ethnicity_chart.svg')
```
